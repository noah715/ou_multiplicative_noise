```julia
import Pkg
Pkg.add("Turing")
Pkg.add("DifferentialEquations")
Pkg.add("Distributions")
Pkg.add("StatsPlots")
Pkg.add("Plots")
Pkg.add("ReverseDiff")
Pkg.add("Memoization")
Pkg.add("FFTW")
```

```julia
using DifferentialEquations
using Plots

μ = 0.0
σ = sqrt(2)
Θ = 1.0
dt = 0.05
W = OrnsteinUhlenbeckProcess(Θ,μ,σ,0.0,1.0)
prob = NoiseProblem(W,(0.0,50.0))
sol = solve(prob;dt=dt)
plot(sol.u)
```

```julia
using Distributions, Random
ou_data = sol.u
noise = rand.(Normal.(0,0.2*sqrt.(abs.(ou_data))))
plot(ou_data.+noise)
plot!(ou_data)
```

```julia
using FFTW
data_fft = fft(ou_data)
noise_fft = fft(ou_data.+noise)

plot(abs.(data_fft), label = "Data")
plot!(abs.(noise_fft), label = "Noise")
savefig("FFT_dt_"*string(dt)*".png")
```

```julia
using Turing, ReverseDiff, Memoization
Turing.setadbackend(:reversediff)
Turing.setrdcache(true)
using Distributions
using LinearAlgebra

# Ornstein-Uhlenbeck process
@model ou(rn,T,delta_t) = begin
    ampl ~ Uniform(0.0,5.0)
    tau ~ Uniform(0.0,5.0)
    
    b = exp(-delta_t/tau)
    
    rn[1] ~ Normal(0,sqrt(ampl))
    
    for i=2:T
        rn[i] ~ Normal(rn[i-1]*b,sqrt(ampl*(1-b^2)))
    end
end

# Ornstein-Uhlenbeck process with added Gaussian noise
@model oupn(rn,T,delta_t,::Type{R}=Vector{Float64}) where {R} = begin
    ampl ~ Uniform(0.0,2.0)
    tau ~ Uniform(0.1,2.0)
    noise_ampl ~ Uniform(0.0,0.5)
    
    b = exp(-delta_t/tau)
    r = R(undef, T)
    
    r[1] ~ Normal(0,sqrt(ampl))
    
    for i=2:T
        r[i] ~ Normal(r[i-1]*b,sqrt(ampl*(1-b^2)))
    end
    rn ~ MvNormal(r,sqrt.(abs.(r)).*noise_ampl)
end
```

```julia
using StatsPlots
@time chnou = sample(ou(ou_data,length(ou_data),0.1), NUTS(0.65), 2000)
# Summarise results (currently requires the master branch from MCMCChains)
describe(chnou)
```

```julia
using StatsPlots
data = ou_data .+ noise
@time chnpn = sample(oupn(data,length(data),0.1), NUTS(0.65), 2000)
describe(chnpn)
```
